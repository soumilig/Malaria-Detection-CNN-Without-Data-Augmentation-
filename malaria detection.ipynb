```python
import keras
from keras.layers import Dense, Conv2D
from keras.layers import MaxPooling2D, Activation, BatchNormalization, Dropout, Flatten
from keras.models import Sequential
from keras import backend as K
from keras import optimizers
```

```python
def CNNb(h, w, c):
    model = Sequential()
    ipshape = (h, w, c)
    chandim = -1
    model.add(Conv2D(64, (3,3), activation='relu', input_shape = ipshape))
    model.add(MaxPooling2D(2, 2))
    model.add(BatchNormalization(axis=chandim))
    model.add(Dropout(0.2))
    
    model.add(Conv2D(64, (3,3), activation='relu', input_shape = ipshape ))
    model.add(MaxPooling2D(2,2))
    model.add(BatchNormalization(axis=chandim))
    model.add(Dropout(0.2))
    
    model.add(Conv2D(32, (3,3), activation='relu', input_shape=ipshape))
    model.add(MaxPooling2D(2,2))
    model.add(BatchNormalization(axis=chandim))
    model.add(Dropout(0.2))
    
    model.add(Flatten())
    
    model.add(Dense(512, activation='relu'))
    model.add(BatchNormalization(axis=chandim))
    model.add(Dropout(0.5))
    model.add(Dense(2, activation='softmax'))
    
    return model
```

```python
emodel = CNNb(h=50,w=50,c=3)
emodel.summary()
```

```python
emodel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
h=emodel.fit(x_train, y_train_2, epochs=20, batch_size=32)
```
